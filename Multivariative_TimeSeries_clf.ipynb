## Creating the sequences

```python
sequences = []
for series_id, group in X_train.groupby("series_id"):
    sequence_features = group[FEATURE_COLUMNS]
    label = y_train[y_train.series_id == series_id].iloc[0].label
    sequences.append((sequence_features, label))
```

```python
train_sequences ,test_sequences = train_test_split(sequences, test_size = 0.2)
print("Number of train sequences: ", len(train_sequences))
print("Number of test sequences: ", len(test_sequences))
```

## Create PyTorch Dataset

```python
class SurfaceDataset(Dataset):
    def __init__(self, sequences):
        super().__init__()
        self.sequences = sequences

    def __len__(self):
        return len(self.sequences)
    
    def __getitem__(self, idx):
        sequence, label = self.sequences[idx]
        return dict(
            sequence = torch.Tensor(sequence.to_numpy()),
            label = torch.tensor(label).long()
        )
```

```python
class SurfaceDataModule(pl.LightningDataModule):
    def __init__(self, train_sequences, test_sequences, batch_size):
        super().__init__()
        self.train_sequences = train_sequences
        self.test_sequences = test_sequences
        self.batch_size = batch_size

    def setup(self, stage=None):
        self.train_dataset = SurfaceDataset(self.train_sequences)
        self.test_dataset = SurfaceDataset(self.test_sequences)

    def train_dataloader(self):
        return DataLoader(
            self.train_dataset,
            batch_size = self.batch_size,
            shuffle = False,
            num_workers = cpu_count()
        )
    
    def val_dataloader(self):
        return DataLoader(
            self.test_dataset,
            batch_size = self.batch_size,
            shuffle = False,
            num_workers = cpu_count()
        )
    
    def test_dataloader(self):
        return DataLoader(
            self.train_dataset,
            batch_size = self.batch_size,
            shuffle = False,
            num_workers = cpu_count()
        )
```

## Model Parameters

```python
N_EPOCHS = 250
BATCH_SIZE = 64

data_module = SurfaceDataModule(train_sequences, test_sequences, BATCH_SIZE)
```

## Model Architecture

```python
class SequenceModel(nn.Module):
    def __init__(self, n_features, n_classes, n_hidden=256, n_layers=3):
        super().__init__()

        self.lstm = nn.LSTM(
            input_size = n_features,
            hidden_size = n_hidden,
            batch_first = True,
            num_layers = n_layers,
            dropout = 0.2
        )

        self.classifier = nn.Linear(n_hidden, n_classes)

    def forward(self, x):
        self.lstm.flatten_parameters()

        _, (hidden, _) = self.lstm(x)
        
        out = hidden[-1]
        return self.classifier(out)
```

```python
class SurfacePredictor(pl.LightningModule):
    def __init__(self, n_features: int, n_classes: int):
        super().__init__()
        self.model = SequenceModel(n_features, n_classes)
        self.criterion = nn.CrossEntropyLoss()

    # 모델 forward
    def forward(self, x, labels=None):
        output = self.model(x)
        loss = 0
        if labels is not None:
            loss = self.criterion(output, labels)
        
        return loss, output
    
    # batch 학습 코드
    def training_step(self, batch, batch_idx):
        sequences = batch['sequence']
        labels = batch['label']

        loss, outputs = self.forward(sequences, labels)
        predictions = torch.argmax(outputs, dim=1)
        step_accuracy = accuracy(predictions, labels)

        self.log("train_loss", loss, prog_bar=True, logger=True)
        self.log("train_accuracy", step_accuracy, prog_bar=True, logger=True)

        return {"loss": loss, "accuracy": step_accuracy}
    
    def validation_step(self, batch, batch_idx):
        sequences = batch['sequence']
        labels = batch['label']

        loss, outputs = self.forward(sequences, labels)
        predictions = torch.argmax(outputs, dim=1)
        step_accuracy = accuracy(predictions, labels)

        self.log("val_loss", loss, prog_bar=True, logger=True)
        self.log("val_accuracy", step_accuracy, prog_bar=True, logger=True)

        return {"loss": loss, "accuracy": step_accuracy}
    
    def configure_optimizers(self):
        return optim.Adam(self.model.parameters(), lr=0.0001)
```

```python
#Model
model = SurfacePredictor(
    n_features=len(FEATURE_COLUMNS),
    n_classes=len(label_encoder.classes_)
)
```

## Training the Model

```python
checkpoint_callback = ModelCheckpoint(
    dirpath="./checkpoints/",
    filename="best-checkpoint",
    save_top_k = 1,
    verbose = True,
    monitor = "val_loss",
    mode = "min"
)

logger = TensorBoardLogger("lightning_logs", name = "surface_predictor")

early_stopping_callback = EarlyStopping(monitor = "val_loss", patience = 10)
```

```python
# Trainer
trainer = pl.Trainer(
    logger = logger,
    checkpoint_callback = checkpoint_callback,
    #callbacks = [early_stopping_callback],
    max_epochs = N_EPOCHS,
    gpus = 1,
    progress_bar_refresh_rate = 30
)
```

```python
trainer.fit(model, data_module)
```

각 모듈과 클래스에 대한 설명 및 사용 이유는 다음과 같습니다</br>

- Dataset: 이는 PyTorch에서 제공하는 기본 Dataset 클래스를 상속받아 구현되었습니다.</br>
데이터셋의 항목에 접근할 때마다 __getitem__ 메소드가 호출되어 해당 항목을 반환합니다.</br>
이를 통해 데이터 로더는 배치 데이터를 얻게 됩니다.</br>

- DataModule (SurfaceDataModule):Lightning에서 제공하는 LightningDataModule을 상속받아 구현되었습니다.</br>
데이터 준비, 분할 및 데이터 로더 설정과 같은 데이터 관련 작업을 캡슐화하는 것을 목적으로 합니다.</br>
이를 통해 데이터를 모델과 분리하며, 다양한 데이터 소스로 쉽게 전환할 수 있습니다.</br>

- SequenceModel:이는 PyTorch의 기본 nn.Module 클래스를 상속받아 구현된 심층 학습 모델입니다.</br>
LSTM 계층과 분류 계층을 포함하고 있습니다.</br>
모델의 전방향 계산을 정의하는 forward 메서드가 있습니다.</br>

- SurfacePredictor:PyTorch Lightning의 LightningModule을 상속받아 구현되었습니다.</br>
LightningModule은 모델의 전방향 계산뿐만 아니라 학습, 검증 및 테스트 로직도 포함하고 있습니다.</br>
또한, 옵티마이저, 학습률 스케줄러 등의 설정도 포함하고 있습니다.</br>
이를 통해 학습과 관련된 모든 코드가 하나의 클래스에 모아져 있어 관리하기 쉽습니다.</br>

## Predictions

```python
#왜 checkpoint 저장이 안됐지???
checkpoint_path = './checkpoints/'

trained_model = SurfacePredictor.load_from_checkpoint(
    checkpoint_path,
    n_feautres = len(FEATURE_COLUMNS),
    n_classes = len(label_encoder.classes_)
)
trained_model.freeze()
```

```python
test_dataset = SurfaceDataset(test_sequences)

# DataLoader를 사용하여 test_dataset 로드
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=cpu_count())

# 모델을 평가 모드로 설정
model.eval()

# GPU 사용 가능한지 확인
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

predictions = []
labels = []

# no_grad 컨텍스트를 사용하여 메모리 사용량을 줄입니다.
with torch.no_grad():
    for batch in tqdm(test_loader):
        sequence = batch["sequence"].to(device)
        label = batch["label"].to(device)

        _, output = model(sequence)
        prediction = torch.argmax(output, dim=1)

        # 결과를 리스트에 추가
        predictions.extend(prediction.cpu().numpy())
        labels.extend(label.cpu().numpy())

# 이제 predictions와 labels는 예측과 라벨을 포함하게 됩니다.
```

```python
print(
    classification_report(
        labels, predictions, target_names=label_encoder.classes_
    )
)
```

```python
def show_confusion_matrix(confusion_matrix):
    hmap = sns.heatmap(confusion_matrix, annot=True, fmt="d", cmap="Blues")
    # hmap.yaxis.set_ticklabels(hmap.yaxis.get_ticklables(), rotation=0, ha="right")
    # hmap.xaxis.set_ticklabels(hmap.xaxis.get_ticklables(), rotation=30, ha="right")
    plt.ylabel("Surface Ground Truth")
    plt.xlabel("Predicted Surface")
    plt.legend()
    
cm = confusion_matrix(labels, predictions)
df_cm = pd.DataFrame(
    cm, 
    index = label_encoder.classes_,
    columns = label_encoder.classes_
)
df_cm
```
